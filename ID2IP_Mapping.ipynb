```python
def portAssign(idList):
  port_map = {}
  base_port = 8000
  last_digit = 0

  for sensor in idList:
    port_map[sensor] = base_port + last_digit
    last_digit += 1
  return port_map

S = [201, 202, 203, 204]
print(portAssign(S))
```
